Getting started with pandas
==========

Pandas consists of high level data structures that are designed to make data analysis easy and intuitive in python. 

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

We will be covering two basic types of data structures supported by pandas:

Systematically, we wil go through the nitty-gritties of each of these data structures in order to see how they work.

Series
------

A Series is a one-dimensional array-like object consisting of data and its associated indices. It can be used as follows:

```python
obj=Series([1, 2, 4, 5])
```

```python
print obj
```

```python
obj.values
```

```python
obj.index
```

You can define your own indices as follows:

```python
obj2=Series([4, 3, -2, 1], index=['a', 'b', 'c', 'd'])
```

```python
print obj2.values
```

```python
print obj2.index
```

```python
print obj2['a']
```

```python
print obj2['d']
```

Boolean indexing works perfectly with the Series data structure.

```python
obj2[obj2 > 0]
```

Like NumPy arrays, Series supports element-wise operations on its elements:

```python
obj2*2
```

```python
np.sqrt(obj2)
```

Even python dictionaries are supported by pandas:

```python
dic={'a':1, 'b':2, 'c':3, 'd':4}
```

```python
obj3=Series(dic)
```

```python
print obj3
```

```python
print obj3.index
```

```python
print obj3.values
```

DataFrame
---------

While Series is a linear, one-dimensional data structure, a DataFrame in contrast is a tabular, spreadsheet-like data structure. A DataFrame has a row and a column index and the row-oriented and column-oriented operations are treated symmetrically. Here, I will use the same example that Wes McKinney uses in his book for purposes of simplicity.

```python
data={'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'], 'year': [2000,2001, 2002, 2001,2002], 'pop': [1.5,1.7, 3.6, 2.4, 2.9]}
```

```python
frame=DataFrame(data)
```

```python
print frame
```

You can specify a particular sequence of columns yourself as follows:

```python
frame=DataFrame(data, columns=['state', 'pop', 'year'])
```

```python
print frame
```

You can retrive data from columns pretty intuitively:

```python
print frame['state']
```

```python
print frame['pop']
```

```python
print frame['year']
```

Columns can be added or modified by assignment.

```python
frame['debt']=24
```

```python
print frame
```

```python
frame['debt']=[12, 13, 14, 15, 16]
```

```python
print frame
```

You can do something similar using Series. Note: If you assign a Series, it will conform to the indexing if the DataFrame *exactly* and any missing values will be assinged NaN.

```python
val = Series([1,2,3], index=[2, 3, 4])
```

```python
frame['debt']=val
```

```python
print frame
```

If you want to delete a column after having inserted it, you can do so as follows:

```python
frame['column']=25
```

```python
print frame
```

```python
print frame.columns
```

```python
del frame['column']
```

```python
print frame
```

```python
print frame.columns
```

Reindexing
----------

The process of changing the indexes of a data structure is called reindexing.

```python
obj=Series([1.2, 2.4, 4.5, 6.7], index=['d', 'b', 'a', 'c'])
```

```python
print obj
```

```python
obj2=obj.reindex(['t', 'd', 'f', 'a', 'b', 'c'])
```

```python
print obj2
```

Since there are no indexes called t or f, they are assigned the NaN value. We can reindex DataFrames in a similar manner.

```python
frame = DataFrame(np.arange(9).reshape((3, 3)), index=['a', 'c', 'd'], columns=['Ohio', 'Texas', 'California'])
```

```python
print frame
```

```python
frame2=frame.reindex([1, 2, 3, 4]) # Every value will be NaN
```

```python
print frame2
```

```python
frame3=frame.reindex(['a','b', 'c','d'])
```

```python
print frame3
```

```python
states = ['Texas', 'Utah', 'California']
frame.reindex(columns=states) #column reindexing
```

```python
frame.reindex(index=['a', 'b', 'c', 'd'], method='ffill', columns=states) # Reindexing both
```

Filtering out missing data
-------------------------

There is clearly some missing data in my definition of the iris dataset. Some lines in the csv files were empty and once they were read by the read_csv() function, they appeared as NaN values. This is a pretty contrived way to incorporate NaN values inside the dataframe, but there are other reasons why we might encounter NaN values in real datasets. For this reason, pandas provides methods of filtering out missing data in the datasets. It can be done as follows:

```python
df
```

```python
df.dropna()
```

```python
df
```

```python
df [df.notnull()] # does the same thing as dropna()
```

```python
cleaned=df.dropna()
```

```python
df.dropna(how='all') # only drops rows that are all NA
```

```python
df
```

```python
df.fillna(0) # fills in the missing data
```
