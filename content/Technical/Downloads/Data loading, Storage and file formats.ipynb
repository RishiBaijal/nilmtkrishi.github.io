You can let pandas assign the default column names or you can assign the names yourself as follows:

```python
pd.read_csv('/Users/rishi/Documents/Master_folder/IIITD/5th_semester/Independent_Project/practice/pydata-book/ch06/iris.csv', names=['col1', 'col2', 'col3', 'col4', 'col5'])
```

As you might have noticed, the file size in this case is relatively large and it becomes inconvenient to scroll up and down on order to get an idea of what the dataset contains. Because of this, python provides you with the ability to read files in pieces. You can do this in the following manner:

```python
df=pd.read_csv('/Users/rishi/Documents/Master_folder/IIITD/5th_semester/Independent_Project/practice/pydata-book/ch06/iris.csv', nrows=5)
```

```python
print df
```

You can also specify a chunk size as a number of rows:

```python
pd.read_csv('/Users/rishi/Documents/Master_folder/IIITD/5th_semester/Independent_Project/practice/pydata-book/ch06/iris.csv', chunksize=10000)
```

Now, you can iterate over parts of the file according to chunksize.

```python

newSer=Series([])
for i in pd.read_csv('/Users/rishi/Documents/Master_folder/IIITD/5th_semester/Independent_Project/practice/pydata-book/ch06/iris.csv', chunksize=10000, names=['col1', 'col2', 'col3', 'col4', 'col5']):
    newSer=newSer.add(i['col1'].value_counts(), fill_value=0)
newSer=newSer.order(ascending=False)
```

```python
print newSer
```
